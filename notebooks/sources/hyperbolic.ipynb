```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from glob import glob
```

```python
f, axes = plt.subplots(len(glob("data/h_psi4/h*.dat")), 1, 
                       figsize=(7,20), 
                       sharex=True)
for i, file in enumerate(glob("data/h_psi4/h*.dat")):
    ax = axes[i]
    waveform = np.genfromtxt(file)
    ax.plot(waveform[:,0], waveform[:,1])
```

```python
f, axes = plt.subplots(len(glob("data/h_psi4/psi*.dat")), 1, 
                       figsize=(7,20), 
                       sharex=True)
for i, file in enumerate(glob("data/h_psi4/psi*.dat")):
    ax = axes[i]
    waveform = np.genfromtxt(file)
    ax.plot(waveform[:,0], waveform[:,1])
```
