```python
import astropy.units as u
import numpy as np

import grasshopper.interferometers as ifo
import grasshopper.sources as sources

import matplotlib.pyplot as plt

import thesis
```

```python
aligo = ifo.AdvancedLIGO()
```

```python
x, y, A,B,H = aligo.skymap(200, 100, psi=1)
```

```python
f = plt.figure(dpi=300)
ax = plt.subplot(111, projection="hammer")
mesh = ax.pcolormesh(y-np.pi,x-np.pi/2., np.sqrt(H**2).T, cmap='viridis', alpha=0.7)
f.colorbar(mesh, orientation='horizontal')
ax.grid(which="major")
f = thesis.thesisify(f);
f.savefig("../../figures/intro/aligo-antenna-pattern.pdf")
```
