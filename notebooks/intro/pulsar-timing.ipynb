```python
import astropy.units as u
import numpy as np

import grasshopper.interferometers as ifo
import grasshopper.sources as sources

import matplotlib.pyplot as plt
#plt.style.use("/home/daniel/thesis/thesis-style.mpl")

import thesis
```

```python
def hellingsdowns(theta):
    first = 0.5 - 0.25 * ((1-np.cos(theta))/2.)
    second = 1.5 * ((1-np.cos(theta))/2.) * np.log((1-np.cos(theta))/2.)
    return first + second
```

```python
theta = np.linspace(0, np.pi, 1000)
```

```python
f, ax = plt.subplots(1,1, dpi=300)
ax.plot(180*(theta/np.pi), hellingsdowns(theta))

ax.set_xlabel("Angular separation between Earth-pulsar baselines [deg]")
ax.set_ylabel("Expected correlation")
ax.set_xlim([0,180])

thesis.thesisify(f);

f.savefig("{}/intro/hellings-downs.{}".format(fig_dir, fig_ext))
```
