```python
data = catalogue[columns][catalogue.index != "GW170817"]

f = plt.figure(figsize=(15,15), dpi=300)

#if not isinstance(additional, pd.DataFrame):
#    # Convert additional to an array
#    additional = pd.DataFrame(additional)

# produce an n x n grid of subplots
gs = gridspec.GridSpec(len(data.columns), len(data.columns),
                       wspace=0.0, hspace=0.0)


# Make a legend for the additional points
#legend_ax = plt.subplot(gs[1,3])
#legend_ax.axis("off")
#legend_ax.grid(None)
#for i, point in enumerate(additional.iterrows()):
#     legend_ax.scatter(0, i, marker="o", c=point[1]['color'], alpha=0.7)
#     legend_ax.text(0.02, i, point[1]['label'], ha="left", va="center", fontdict=lato)
#legend_ax.set_xlim([-0.05, 0.3])

for i, parameter in enumerate(data.columns):
    for j, j_parameter in enumerate(data.columns):

        if i >= j:
            # Don't produce a plot for combinations above the
            # diagonal.
            continue
        else:
            # Produce a subplot for combinations on or below
            # the diagonal
            ax = plt.subplot(gs[j, i])
            ax.grid(None)

        if i == j:
            pass
        #    # This is the on-diagonal case, which we'll just skip for
        #    # now. Ideally will want to insert a histogram here.
        #    ax.hist(map(lambda x: x['best'], data[parameter]), histtype="step")
        #    ax.yaxis.tick_right()
        #    ax.set_yticks(ax.get_yticks()[1:-1])
        else:

            # Add in the 'additional' points
            #ax.scatter(additional[parameter], additional[j_parameter], marker="o", c=additional['color'], alpha=0.7)


            # Produce a scatter plot of the waveforms for this
            # combination
            #ax.scatter(map(lambda x: x['best'], data[parameter]),
            #           map(lambda x: x['best'], data[j_parameter]),
            #           marker=".")
            
            def return_bounds(x):
                if "err" in x:
                    lims = np.array([0,0])
                    bounds = np.array([0,0])
                    if x["err"][5:] != "bound":
                        bounds =  np.array(x['err']) 
                    elif x["err"] == "upperbound":
                        bounds = np.array([0, 1])
                        lims = np.array([0, 1])
                    elif x["err"] == "lowerbound":
                        lims = np.array([1, 0])
                        bounds = np.array([1, 0])
                    bounds[0] *= -1
                    return bounds, lims
                else:
                    return np.array([0,0]), np.array([0,0]), np.array([0,0])
            
            xerr = np.array(map(return_bounds , data[parameter]))[:,0]
            xlim =  np.array(map(return_bounds , data[parameter]))[:,1]
            yerr = np.array(map(return_bounds, data[j_parameter]))[:,0]
            ylim = np.array(map(return_bounds, data[j_parameter]))[:,1]

            ax.errorbar(x=map(lambda x: x['best'], data[parameter]),
                    y=map(lambda x: x['best'], data[j_parameter]),
                    xerr=xerr.T,
                    xlolims=xlim[:,0],
                        xuplims=xlim[:,1],
                        lolims=ylim[:,0],
                        uplims=ylim[:,1],
                    yerr=yerr.T,
                    ls="", marker=".")
            
            
        if j == len(data.columns) - 1:
            ax.set_xlabel(parameter.replace("_", " "))
        else:
            ax.set_xticks([])

        if i == 0:
            ax.set_ylabel(j_parameter.replace("_", " "))
        elif not i == j:
            ax.set_yticks([])

f.tight_layout()
```
