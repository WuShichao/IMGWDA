```python
fig_dir = "../../figures"
fig_ext = "pdf"
```

```python
import astropy.units as u
import numpy as np

import grasshopper.interferometers as ifo
import grasshopper.sources as sources
import grasshopper.timingarray as ta

import matplotlib.pyplot as plt
#plt.style.use("/home/daniel/thesis/thesis-style.mpl")

import thesis
```

```python
ipta = ta.IPTA()
```

```python
fig, ax = plt.subplots(1,1, dpi=300)
ipta_line = ipta.plot(ax)#, frequencies=np.logspace(-10,-5))
ipta_line[0].set_xdata(ipta_line[0].get_xdata()*u.hertz)
thesis.labelLine(ipta_line[0], label="IPTA", x=1.5e-7, yshift=1.25,  fontsize=8)
thesis.thesisify(fig)
fig.savefig("{}/intro/ipta-asd.{}".format(fig_dir, fig_ext), dpi=300)
```
